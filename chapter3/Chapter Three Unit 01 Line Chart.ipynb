```python
a , b = plt.subplots(2, 2)
# print(type(a))
# print(type(b))
# type(a) = matplotlib.figure
# type(b) = numpy.ndarray

b[0,0].plot(stock_data_dateindex['AAPL'])
b[0,1].plot(stock_data_dateindex['MSFT'])
b[1,0].plot(stock_data_dateindex['IBM'])
b[1,1].plot(stock_data_dateindex['CSCO'])

a.suptitle('Stock Trends')
plt.show()
```

```python
fig , axes = plt.subplots(2, 2 , sharey=True)

axes[0,0].plot(stock_data_dateindex['AAPL'])
axes[0,0].set_title('Apple Stock')

axes[0,1].plot(stock_data_dateindex['MSFT'])
axes[0,1].set_title('microsft Stock')

axes[1,0].plot(stock_data_dateindex['IBM'])
axes[1,0].set_title('IBM Stock')

axes[1,1].plot(stock_data_dateindex['CSCO'])
axes[1,1].set_title('Cisco Stock')

fig.suptitle('Stock Trends')
plt.tight_layout()
plt.show()
```

```python
# Another way to create subplots
plt.subplots(2,2, sharey='all',sharex='all')

plt.subplot(2,2,1)
plt.plot(stock_data['Date'],stock_data['AAPL'])
plt.title('Apple Stock')

plt.subplot(2,2,2)
plt.plot(stock_data['Date'],stock_data['MSFT'])
plt.title('Microsoft Stock')

plt.subplot(2,2,3)
plt.plot(stock_data['Date'],stock_data['CSCO'])
plt.title('Cisco Stock')

plt.subplot(2,2,4)
plt.plot(stock_data['Date'],stock_data['IBM'])
plt.title('IBM Stock')
  
plt.tight_layout()
plt.show()
```
