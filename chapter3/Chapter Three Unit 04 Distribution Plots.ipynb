## Histograms

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv('cars.csv')
cars.info()
```

```python
sns.displot(data=cars, x='city_miles_per_galloon');
```

```python
sns.displot(data=cars, x='city_miles_per_galloon', binwidth=0.5);
```

```python
sns.displot(data=cars, x='city_miles_per_galloon', kde=True, color='green');
```

```python
sns.displot(data=cars, x='highway_miles_per_Gallon', kde=True);
```

```python
# Figure-Level Function
# sns.distplot
sns.displot(data=cars,
            x='city_miles_per_galloon',
            kde=True,
            color='darkblue',
            label='City')
sns.displot(data=cars,
            x='highway_miles_per_Gallon',
            kde=True,
            color='darkred',
            label='Highway')
plt.legend()
plt.xlabel('Miles')
plt.show()
```

```python
# Axis-Level Function
sns.histplot(data=cars,
             x='city_miles_per_galloon',
             kde=True,
             color='darkblue',
             label='City')
sns.histplot(data=cars,
             x='highway_miles_per_Gallon',
             kde=True,
             color='darkred',
             label='Highway')
plt.legend()
plt.xlabel('Miles')
plt.show()
```

```python
sns.displot(data=cars,
            x='city_miles_per_galloon',
            hue='cylenders',
            palette='bright');
```

```python
cars_common = cars[cars.cylenders.isin([4, 8, 6])]
cars_common.head()
```

```python
sns.displot(data=cars_common,
            x='city_miles_per_galloon',
            hue='cylenders',
            palette='bright');
```

```python
sns.displot(data=cars_common,
            x='city_miles_per_galloon',
            hue='cylenders',
            palette='bright',
            element='step');
```

```python
sns.displot(data=cars_common,
            x='city_miles_per_galloon',
            col='cylenders',
            hue='cylenders',
            palette='bright');
```

## Kernel density estimation

```python
sns.displot(data=cars_common,
            x='city_miles_per_galloon',
            hue='cylenders',
            palette='bright',
            kind='kde');
```

```python
sns.displot(data=cars_common,
            x='city_miles_per_galloon',
            hue='cylenders',
            palette='bright',
            kind='kde',
            fill=True);
```

```python
# sns.kdeplot(data=cars_common,
#     x='city_miles_per_galloon',
#     hue='cylenders',
#     palette='bright')

sns.displot(
    data=cars_common,
    x='city_miles_per_galloon',
    col='cylenders',
    hue='cylenders',
    palette='bright',
    kind = 'kde');
```

## Empirical cumulative distributions

```python
# ecdf = Emprical Cumulative Distribution Function
# sns.displot(kind='ecdf')

graph = sns.ecdfplot(data=cars_common,
                     x='city_miles_per_galloon',
                     hue='cylenders',
                     palette='bright');
```

```python
graph = sns.ecdfplot(data=cars_common,
                     x='city_miles_per_galloon',
                     hue='cylenders',
                     palette='bright');

graph.axhline(0.5);
# Find Median
# Find Mean
```

## Box Plot

```python
plt.figure(figsize=(9, 5))
# sns.boxenplot(x='cylenders', y='city_miles_per_galloon', data=cars_common);
sns.boxplot(x='cylenders', y='city_miles_per_galloon', data=cars_common);
```

```python
plt.figure(figsize=(14, 6))
sns.swarmplot(x='cylenders',
              y='city_miles_per_galloon',
              data=cars_common);
```

```python
plt.figure(figsize=(9, 5))
sns.boxplot(x='cylenders',
            y='city_miles_per_galloon',
            data=cars_common,
            showfliers=False);
```

```python
ordered_cyl = cars_common.groupby('cylenders'
                                  ).city_miles_per_galloon.median().sort_values().index
# ordered_cyl
plt.figure(figsize=(9, 5))
sns.boxplot(x='cylenders',
            y='city_miles_per_galloon',
            data=cars_common,
            order=ordered_cyl,
            showfliers=False);
```

```python
plt.figure(figsize=(9, 5))
# Figure-Level Function
sns.catplot(x='cylenders', y='city_miles_per_galloon', data=cars_common , kind='box')
sns.catplot(x='cylenders', y='highway_miles_per_Gallon', data=cars_common, kind='box')
plt.show()
```

```python
plt.figure(figsize=(9, 5))
# Axis-Level Function
sns.boxplot(x='cylenders', y='city_miles_per_galloon', data=cars_common )
sns.boxplot(x='cylenders', y='highway_miles_per_Gallon', data=cars_common)
plt.show()
```

```python
plt.subplots(1,2,sharey=True)

plt.subplot(1, 2, 1)
sns.boxplot(x='cylenders', y='city_miles_per_galloon', data=cars_common)
plt.title('City')

plt.subplot(1, 2, 2)
sns.boxplot(x='cylenders', y='highway_miles_per_Gallon', data=cars_common)
plt.title('Highway')
plt.show()
```

## Joint and Marginal distributions

```python
g = sns.JointGrid(data=cars_common, x='city_miles_per_galloon', y='Price')
g.plot_joint(sns.scatterplot)
g.plot_marginals(sns.boxplot, color='purple');
sns.set()
```
