## Annotation

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=cars_data, 
                x='city_miles_per_galloon',
                y='Price'
                )
sns.lineplot(data=cars_data, 
            x='city_miles_per_galloon', 
            y= cars_data.Price.mean(), 
            color='g',  
            ci=None
            );
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=cars_data, 
                x='city_miles_per_galloon',
                y='Price'
                )
plt.axhline(y= cars_data.Price.mean(),color='g');
```

```python
p1 = sns.scatterplot(data=cars_data, 
                     x='city_miles_per_galloon',
                     y='Price'
                    )
p2 = sns.lineplot(data=cars_data, 
                  x='city_miles_per_galloon', 
                  y= cars_data.Price.mean(), 
                  color='g',  
                  ci=None
                 );
```

```python
sns.scatterplot(data=cars_data, 
                 x='city_miles_per_galloon',
                 y='Price')
sns.lineplot(data=cars_data, 
                  x='city_miles_per_galloon', 
                  y=cars_data.Price.mean(), 
                  color='g' , 
                  ci=None)
plt.annotate(text='Average Price', 
            xy=(30,cars_data.Price.mean()+3000),
            color='brown');
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=cars_data, x='city_miles_per_galloon',y='Price')
sns.lineplot(data=cars_data, x='city_miles_per_galloon', y=cars_data.Price.mean(), color='g' , ci=None)
plt.annotate(text='Average Price', xy=(30,cars_data.Price.mean()+3000),color='brown')

plt.annotate(text="Outlier",color='purple',
            xy = (18, 190000),
            xytext = (22, 175000),
            arrowprops = dict(arrowstyle = "->",
                            connectionstyle = "arc3",
                            color='purple')
           );
```

https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.FancyArrowPatch.html#matplotlib.patches.FancyArrowPatch

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=cars_data, x='city_miles_per_galloon', y='Price')
sns.lineplot(data=cars_data,x='city_miles_per_galloon',y=cars_data.Price.mean(),color='g',ci=None)

plt.annotate('Average Price', xy=(30, cars_data.Price.mean() + 3000), color='brown')
plt.annotate("Outlier",color='purple',
            xy=(18, 190000),xytext=(22, 175000),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3",
                            color='purple'))

point1 = [10, 1000]
point2 = [60, 175000]
x_values = [point1[0], point2[0]]
y_values = [point1[1], point2[1]]
plt.plot(x_values, y_values, 'darkred', linestyle="--");
```
