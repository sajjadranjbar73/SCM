## 1. Merge

<img src='Join-Types.png'  height="400px" width="400px" style = "float:left"/>

```python
import pandas as pd
purchase = pd.read_excel('Purchase.xlsx')
profile  = pd.read_excel('Profile.xlsx')
```

```python
purchase
```

```python
profile
```

```python
# By defualt pd.merge() defaults ---> ON = same columns & HOW = inner

merged_data = pd.merge(purchase,profile)
merged_data

#### Another way:
# merged_data = purchase.merge(profile)
```

```python
# merge on specific column

merged_data = pd.merge(purchase,profile, on ='Customer Name')
merged_data
```

```python
# Assign suffixes to same column names on different DataFrames

merged_data = pd.merge(purchase,
                       profile, 
                       on ='Customer Name',
                       suffixes=('_Purchase', '_Profile')
                      )
merged_data
```

```python
# Choosing desired columns from DataFrames

merged_data = pd.merge(purchase,
                       profile[['Customer Name','Age']], 
                       on ='Customer Name',
                       suffixes=('_Purchase', '_Profile')
                      )
merged_data
```

```python
merged_data = pd.merge(purchase[['Customer Name','Quantity']],
                       profile[['Customer Name','Age']], 
                       on ='Customer Name',
                       suffixes=('_Purchase', '_Profile')
                      )
merged_data
```

### 1.1. Other merge types

```python
# Outer merge

merged_data = pd.merge(purchase,
                       profile,
                       on ='Customer Name',
                       suffixes=('_Purchase', '_Profile'),
                       how ='outer'
                      )
merged_data
```

```python
# Left merge

merged_data = pd.merge(purchase,
                       profile,
                       on ='Customer Name',
                       suffixes=('_Purchase', '_Profile'),
                       how ='left'
                      )
merged_data
```

```python
# Right merge

merged_data = pd.merge(purchase,
                       profile,
                       on ='Customer Name',
                       suffixes=('_Purchase', '_Profile'),
                       how ='right'
                      )
merged_data
```

### 2.1. Merge on different column names

```python
invoice_total = pd.read_excel('Invoice.xlsx','Invoice_Total')
invoice_total
```

```python
invoice_item  = pd.read_excel('Invoice.xlsx','Invoice_Item')
invoice_item
```

```python
merge_invoice = pd.merge(invoice_item,
                         invoice_total,
                         left_on = 'Invoice_ID',
                         right_on = 'Invoice_Num'                        
                        )
merge_invoice
```

```python
merge_invoice = pd.merge(invoice_item ,
                         invoice_total[['Invoice_Num','Date']] ,
                         left_on = 'Invoice_ID',
                         right_on = 'Invoice_Num'                        
                        )
merge_invoice
```

```python
merge_invoice.drop(columns='Invoice_Num',inplace=True)
```

```python
merge_invoice
```

```python
merge_invoice['Net_Price'] = merge_invoice.Quantity * merge_invoice.Price
merge_invoice
```

```python
merge_invoice.groupby('Date')['Net_Price'].sum().reset_index()    
```
