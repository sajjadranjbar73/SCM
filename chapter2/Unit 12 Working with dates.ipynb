```python
grp_customer = retail.groupby('Customer ID').agg(
    first_date=('InvoiceDate', 'min'),
    last_date=('InvoiceDate', 'max')).reset_index()
```

```python
grp_customer.head()
```

```python
grp_customer['interval'] = grp_customer['last_date'] - grp_customer['first_date']
grp_customer.head()
```

```python
grp_customer.info()
```

```python
# works only for timedelta 
grp_customer.interval.dt.components
```

```python
# df.timedelta.dt.components is a DataFrame
grp_customer.interval.dt.components[['days','hours']]
```

```python
grp_customer['interval_days'] = grp_customer.interval.dt.components['days']
grp_customer.head()
```

```python
current_time = datetime.now()
# current_time
type(current_time)
```

```python
grp_customer['far_from_now'] = current_time - grp_customer['last_date']
```

```python
grp_customer.head()
```

```python
grp_customer.info()
```

```python
grp_customer['from_now_days'] = grp_customer.far_from_now.dt.components['days']
grp_customer.head()
```
